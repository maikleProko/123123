## Основные теоретические положения
Def 1.1 Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими свойствами:

1) $ex = x$

2) $g(hx) = (gh)x$

где $gx$ - это образ пары $(g,x)$.

Def 1.2 Пусть $X$ - множество. Множество всех биективных функций с $X \to X$ с операцией композиции называется симметрической группой на множестве $X$ и обозначается как $S_{X}$

Def 1.3 Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

Def 1.4 Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента g обозначается через $X^{g}$.

Th 1 (Burnside) Количество орбит действия группы $G$ на множестве $X$ равно:

$\frac{1}{|G|}\sum_{g \in G} |X^{g}|$

Lemma 1 Предположим, что $\sigma$ является циклом длины $k$, а $НОД(k,n) = d$. Тогда перестановка $\sigma^{n}$ является произведением d циклов длины $\frac{k}{d}$.

Def 1.5 Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^{I}$.

Lemma 2 Пусть $G$ группа, действующая на множестве $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$.

Lemma 3 Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1)$gf = f$

2)$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$

3)Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$ (в терминологии задачи о бусах, это означает, что все места из одного цикла покрашены в одинаковый цвет)

Corollary 1 Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $|C^{c(\varphi(g))}|$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).
