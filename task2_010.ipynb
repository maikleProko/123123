## Основные теоретические положения

**Определение 1:** 

Орбита элемента $x \in X$ под действием $G$ - это множество $Gx = \{gx | g \in G\}$, где численность элементов в ней предполагает её длину.

При этом, две орбиты не могут пересекаться, либо они должны совпадать, из чего  множество $A$ разбивается в дизъюнктивное объединение орбит.

**Определение 2:** 

Cимметрическая группа на некотором множестве $A$ - это множество всех биективных функций с $X \to X$ с операцией композиции.

Обозначается следующим образом: $S_{X}$

**Определение 3:** 

Раскраска некоторого множества $A$ - это функция отображения из $A$ в некоторое множество цветов $C$.

Обозначается следующим образом: $C^{I}$

**Определение 4:** 

Предположим $G$ - это группа, имеющая нейтральный элемент $e$, а $A$ - множество. 

Пусть задана операция $G \times X \to X$, со свойствами: $ex = x$; $g(hx) = (gh)x$  (при этом $gx$ является образом пары $(g,x)$)

Тогда $G$ действует на $A$.

**Определение 5:** 

Неподвижные точки элемента $g \in G$ это $x \in X$, для которых выполняется:

$gx = x$.

Обозначается следующим образом: $X^{g}$.

**Лемма Бернсайда:** 

Количество орбит действия группы $G$ на множестве $X$ равно:

$N=\frac{1}{|G|}\sum_{g \in G} |X^{g}|$, где $X^{g}$ - это неподвижные точки элемента $g \in G$

Для того чтобы посчитать количество способов раскраски граней тетраэдра, воспользуемся леммой Бернсайда:

**Функция, отображающая графически раскраску по заданному номеру.**

Была реализована функция, отображающая графически раскраску по заданному номеру:

```python
import math
import random
from sage.plot.plot3d.shapes import Text

def gen_col(var):
    ret_str=""
    if(var==1):
        return "red"
    if(var==2):
        return "green"
    if(var==3):
        return "blue"
    if(var==4):
        return "white"
    if(var==5):
        return "black"
    else:
        return "empty"

def paint_tetr(col):
    x=[]
    y=[]
    z=[]
    for i in range(4):
        x.append(0)
        y.append(0)
        z.append(0)

    z[0]=z[1]=0
    z[2]=2
    z[3]=2+1

    x[0]=2
    x[1]=-x[0]
    x[2]=x[0]+1
    x[3]=x[2]+2

    y[0]=1
    y[1]=2
    y[2]=3
    y[3]=2


    Pol1=polygon3d([(x[0],y[0],z[0]), (x[1],y[1],z[1]), (x[2],y[2],z[2])], color = gen_col(col[0]))
    Pol2=polygon3d([(x[0],y[0],z[0]), (x[1],y[1],z[1]), (x[3],y[3],z[3])], color = gen_col(col[1]))
    Pol3=polygon3d([(x[3],y[3],z[3]), (x[1],y[1],z[1]), (x[2],y[2],z[2])], color = gen_col(col[2]))
    Pol4=polygon3d([(x[0],y[0],z[0]), (x[3],y[3],z[3]), (x[2],y[2],z[2])], color = gen_col(col[3]))

    lgnd = Text("1", color = 'black').translate(0.3,0.5,-0.1) + Text("2", color = 'black').translate(0.5, 0.2, 0.3) + Text("3", color = 'black').translate(0.5, 0.7, 0.3) + Text("4", color = 'black').translate(0,0.5,0.5)
    return Pol1 + Pol2 + Pol3 + Pol4 + lgnd

col=[]
for i in range(4):
    col.append(random.randint(1,5))
graph = paint_tetr(col)
graph.show()
```
